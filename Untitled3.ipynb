```python
from ultralytics import YOLO
import cv2
from pathlib import Path
from IPython.display import display, Image as IPyImage

# === Step 1: Load your trained YOLOv8 model ===
model = YOLO('runs/detect/train/weights/best.pt')

# === Step 2: Inference on your image ===
image_path = '2pixapothole.jpg'
results = model(image_path, save=True, conf=0.4)  # set confidence threshold

# === Step 3: Get path to output image with bounding boxes ===
output_dir = Path(results[0].save_dir)  # runs/detect/predict
output_img_path = output_dir / Path(image_path).name  # e.g., runs/detect/predict/2pixapothole.jpg

# === Step 4: Display the output image ===
display(IPyImage(filename=str(output_img_path)))

# === Optional: Show with OpenCV window (not visible in Colab, useful for local run) ===
# output_img = cv2.imread(str(output_img_path))
# cv2.imshow("Detections", output_img)
# cv2.waitKey(0)
# cv2.destroyAllWindows()

```
